```python
# ---------------------------------------------------------
# [Step 1 & 2] Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨, Í≤ÄÏ¶ù Î∞è Î≥ÄÏàò ÏÑ†ÌÉù (Complete Version)
# ---------------------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LassoCV
from sklearn.feature_selection import mutual_info_regression

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï
import matplotlib.font_manager as fm
import os

font_candidates = [
    "/usr/share/fonts/truetype/nanum/NanumGothic.ttf",
    "/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf"
]
font_path = None
for path in font_candidates:
    if os.path.exists(path):
        font_path = path
        break
if font_path:
    font_prop = fm.FontProperties(fname=font_path)
    plt.rcParams['font.family'] = font_prop.get_name()
    plt.rcParams['axes.unicode_minus'] = False
    print("‚úÖ ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï ÏôÑÎ£å")
else:
    print("‚ö†Ô∏è ÌïúÍ∏Ä Ìè∞Ìä∏ ÎØ∏Î∞úÍ≤¨ (ÏòÅÏñ¥Î°ú Ï∂úÎ†•Îê©ÎãàÎã§)")

# ---------------------------------------------------------
# 1. Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è ÏòÅÎ¨∏Î™Ö Îß§Ìïë
# ---------------------------------------------------------
file_path = '2000-2025 Ï∑®Ìï©.xlsx'

try:
    df = pd.read_excel(file_path)
    print("‚úÖ ÌååÏùº Î°úÎìú ÏÑ±Í≥µ")
except FileNotFoundError:
    raise FileNotFoundError("ÌååÏùºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§. ÏóÖÎ°úÎìú ÏÉÅÌÉúÎ•º ÌôïÏù∏ÌïòÏÑ∏Ïöî.")

df['observation_date'] = pd.to_datetime(df['observation_date'])
df = df.set_index('observation_date').sort_index()

col_map = {
    'USD_KRW ÏõîÎ≥Ñ': 'FX',
    'VIXCLS': 'VIX',
    'WTI': 'WTI',
    'USD_JPY': 'USD_JPY',
    'S&P 500': 'SP500',
    'M2SL': 'M2_US',
    'M2 ÏÉÅÌíàÎ≥Ñ Íµ¨ÏÑ±': 'M2_KR',
    'ÌïúÎØ∏2ÎÖÑ Íµ≠Ï±ÑÏä§ÌîÑÎ†àÎìú': 'Spread_KR_US',
    'ÌïúÍµ≠ÏùÄÌñâ Í∏∞Ï§ÄÍ∏àÎ¶¨ Î∞è Ïó¨ÏàòÏã†Í∏àÎ¶¨': 'BaseRate_KR',
    'ÏΩîÏä§ÌîºÏßÄÏàò Í≥ºÍ±∞ Îç∞Ïù¥ÌÑ∞': 'KOSPI',
    'ÏÜåÎπÑÏûêÎ¨ºÍ∞ÄÏßÄÏàò': 'CPI_KR',
    'ÎØ∏Íµ≠ PPI': 'PPI_US',
    'ÎØ∏Íµ≠ NFP': 'NFP_US',
    'ÎØ∏Íµ≠ CPI': 'CPI_US',
    'ÎØ∏Íµ≠COPRECPI': 'CoreCPI_US',
    'ÎØ∏Íµ≠ Í∏∞Ï§ÄÍ∏àÎ¶¨': 'BaseRate_US',
    'ÎØ∏Íµ≠ 10ÎÖÑÎ¨º Í∏àÎ¶¨': 'Treasury_10Y_US',
    'Îã¨Îü¨Ïù∏Îç±Ïä§': 'DXY'
}
df = df.rename(columns=col_map)

# ---------------------------------------------------------
# 2. Ï†ïÏÉÅÏÑ± Î≥ÄÌôò (Stationary Transformation)
# ---------------------------------------------------------
print("\nüîÑ Î≥ÄÏàò Î≥ÄÌôò Ï§ë (Log Diff or Diff)...")
df_trans = pd.DataFrame()
df_trans['FX_Level'] = df['FX'] # ÏõêÎ≥∏ Î†àÎ≤® Ï†ÄÏû• (Î≥µÏõêÏö©)

# Î≥ÄÌôò Î°úÏßÅ Ï†ÅÏö©
for col in df.columns:
    if col == 'FX':
        # ÌÉÄÍ≤üÏùÄ Î°úÍ∑∏ Ï∞®Î∂Ñ (ÏàòÏùµÎ•†)
        df_trans['dlog_FX'] = np.log(df['FX']) - np.log(df['FX'].shift(1))
    else:
        # 0 Ïù¥Ìïò Í∞íÏù¥ ÏûàÍ±∞ÎÇò Í∏àÎ¶¨/Ïä§ÌîÑÎ†àÎìúÎ•òÎäî Îã®Ïàú Ï∞®Î∂Ñ(diff)
        if (df[col] <= 0).any():
            df_trans[f'diff_{col}'] = df[col].diff()
        else:
            # Í∑∏ Ïô∏(ÏßÄÏàò, Í∞ÄÍ≤© Îì±)Îäî Î°úÍ∑∏ Ï∞®Î∂Ñ(dlog)
            df_trans[f'dlog_{col}'] = np.log(df[col]) - np.log(df[col].shift(1))

# Í≤∞Ï∏°Ïπò Ï†úÍ±∞
df_trans = df_trans.dropna()

# ---------------------------------------------------------
# [Í≤ÄÏ¶ù 1] ADF Test (Îã®ÏúÑÍ∑º Í≤ÄÏ†ï)
# ---------------------------------------------------------
print("\nüîç [Í≤ÄÏ¶ù 1] ADF Test (Ï†ïÏÉÅÏÑ± ÌôïÏù∏)")
stationary_cols = []
cols_to_check = [c for c in df_trans.columns if c != 'FX_Level']

print(f"{'Variable':<25} {'p-value':<10} {'Result'}")
print("-" * 50)

for col in cols_to_check:
    series = df_trans[col].replace([np.inf, -np.inf], np.nan).dropna()
    res = adfuller(series)
    p_value = res[1]

    if p_value < 0.05:
        stationary_cols.append(col)
        print(f"{col:<25} {p_value:.4f}     Pass")
    else:
        print(f"{col:<25} {p_value:.4f}     Fail (Excluding...)")

print("-" * 50)
print(f"‚úÖ Ï¥ù {len(cols_to_check)}Í∞ú Ï§ë {len(stationary_cols)}Í∞ú Î≥ÄÏàòÍ∞Ä Ï†ïÏÉÅÏÑ± ÌôïÎ≥¥Îê®.")

# ---------------------------------------------------------
# 3. ÌÉÄÍ≤ü ÏÉùÏÑ± (Shift -1)
# ---------------------------------------------------------
# Ï†ïÏÉÅÏÑ±Ïù¥ ÌôïÎ≥¥Îêú Î≥ÄÏàòÎì§Î°úÎßå Îç∞Ïù¥ÌÑ∞ÏÖã Íµ¨ÏÑ±
df_final = df_trans[stationary_cols + ['FX_Level']].copy()
df_final['Target_Next_Month'] = df_final['dlog_FX'].shift(-1)
df_final_train = df_final.dropna() # ÎßàÏßÄÎßâ Ìñâ(ÌÉÄÍ≤ü ÏóÜÏùå) Ï†úÍ±∞

# ---------------------------------------------------------
# 4. Î≥ÄÏàò Ï§ëÏöîÎèÑ ÏÇ∞Ï∂ú (Feature Importance)
# ---------------------------------------------------------
print("\nüìä Î≥ÄÏàò Ï§ëÏöîÎèÑ Î∂ÑÏÑù ÏãúÏûë...")
X = df_final_train.drop(columns=['Target_Next_Month', 'FX_Level'])
y = df_final_train['Target_Next_Month']
feature_names = X.columns.tolist()

results = pd.DataFrame(index=feature_names)

# 1) ÏÉÅÍ¥ÄÍ≥ÑÏàò
results['Corr'] = df_final_train.corr()['Target_Next_Month'][feature_names].abs()

# 2) Mutual Information
results['MI'] = mutual_info_regression(X, y, random_state=42)

# 3) Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X, y)
results['RF'] = rf.feature_importances_

# 4) Lasso
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
lasso = LassoCV(cv=5, random_state=42).fit(X_scaled, y)
results['Lasso'] = np.abs(lasso.coef_)

# ÌÜµÌï© Ï†êÏàò (MinMax Scaling ÌõÑ ÌèâÍ∑†)
mm = MinMaxScaler()
results_sc = pd.DataFrame(mm.fit_transform(results), index=results.index, columns=results.columns)
results['Total_Score'] = results_sc.mean(axis=1)
results = results.sort_values('Total_Score', ascending=False)

print("   -> Ï§ëÏöîÎèÑ ÏÇ∞Ï∂ú ÏôÑÎ£å.")

# ---------------------------------------------------------
# [Í≤ÄÏ¶ù 2] Îã§Ï§ëÍ≥µÏÑ†ÏÑ± Í∏∞Î∞ò Î≥ÄÏàò ÌïÑÌÑ∞ÎßÅ (Smart Selection)
# ---------------------------------------------------------
# Ï§ëÏöîÎèÑ ÏÉÅÏúÑ 15Í∞ú Î≥ÄÏàò Ï§ë, ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ 0.9 Ïù¥ÏÉÅÏù¥Î©¥ Ï†êÏàò ÎÇÆÏùÄÍ±∞ ÌÉàÎùΩÏãúÌÇ¥
print("\nüîç [Í≤ÄÏ¶ù 2] Îã§Ï§ëÍ≥µÏÑ†ÏÑ±(Multicollinearity) ÌïÑÌÑ∞ÎßÅ")
top_candidates = results.head(15).index.tolist()
final_selected = []

# Ï§ëÏöîÎèÑ ÏàúÏúºÎ°ú ÌïòÎÇòÏî© ÎÑ£ÏúºÎ©¥ÏÑú, Ïù¥ÎØ∏ ÎΩëÌûå Ïï†Îì§Í≥º ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Ï≤¥ÌÅ¨
correlation_matrix = X[top_candidates].corr().abs()

print(f"   -> Ï¥àÍ∏∞ ÌõÑÎ≥¥({len(top_candidates)}Í∞ú): {top_candidates}")

for candidate in top_candidates:
    is_correlated = False
    for selected in final_selected:
        corr_val = correlation_matrix.loc[candidate, selected]
        if corr_val > 0.9: # ÏûÑÍ≥ÑÍ∞í 0.9
            print(f"      üö´ ÌÉàÎùΩ: {candidate} (Ïú†ÏÇ¨Î≥ÄÏàò {selected}ÏôÄ ÏÉÅÍ¥ÄÍ≥ÑÏàò {corr_val:.2f})")
            is_correlated = True
            break

    if not is_correlated:
        final_selected.append(candidate)

# ÏµúÏ¢Ö 10Í∞úÍπåÏßÄÎßå Ïú†ÏßÄ (ÎÑàÎ¨¥ ÎßéÏúºÎ©¥ Ïò§Î≤ÑÌîºÌåÖ)
final_selected = final_selected[:10]

print("-" * 50)
print(f"‚úÖ ÏµúÏ¢Ö ÌôïÏ†ïÎêú ÏÑ§Î™ÖÎ≥ÄÏàò ({len(final_selected)}Í∞ú):")
print(final_selected)
print("-" * 50)

# ÏãúÍ∞ÅÌôî (ÏÑ†ÌÉùÎêú Î≥ÄÏàòÎì§Ïùò Ï§ëÏöîÎèÑ)
plt.figure(figsize=(10, 6))
sns.barplot(x=results.loc[final_selected, 'Total_Score'], y=final_selected, palette='viridis')
plt.title('Final Selected Features Importance')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# 5. ÏµúÏ¢Ö Îç∞Ïù¥ÌÑ∞ÏÖã Ï†ÄÏû• (Step 3Ïö©)
# ---------------------------------------------------------
# Step 3ÏóêÏÑú Ïì∏ Î≥ÄÏàò Î¶¨Ïä§Ìä∏ÏôÄ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ
# Ï†ÑÏó≠ Î≥ÄÏàò 'df_final_train', 'final_selected'Í∞Ä Î©îÎ™®Î¶¨Ïóê ÎÇ®ÏäµÎãàÎã§.
print("\nüéâ Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ ÏôÑÎ£å! Îã§Ïùå Îã®Í≥Ñ(Î™®Îç∏ÎßÅ)Î°ú ÎÑòÏñ¥Í∞à Ï§ÄÎπÑÍ∞Ä ÎêòÏóàÏäµÎãàÎã§.")
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import random
import os
import itertools

from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, Attention, GlobalAveragePooling1D
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

try:
    import optuna
except ImportError:
    !pip install optuna
    import optuna

# 1. ÌôòÍ≤Ω ÏÑ§Ï†ï (Reproducibility)
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

print(f"TensorFlow Version: {tf.__version__}")

# 2. Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏ (Step 1&2 Í≤∞Í≥ºÎ¨º)
if 'df_final_train' not in locals() or 'final_selected' not in locals():
    raise ValueError("‚ö†Ô∏è Step 1_2 ÏΩîÎìúÎ•º Î®ºÏ†Ä Ïã§ÌñâÌïòÏó¨ Îç∞Ïù¥ÌÑ∞(df_final_train, final_selected)Î•º Ï§ÄÎπÑÌï¥Ï£ºÏÑ∏Ïöî.")

print(f"\n‚úÖ ÌïôÏäµ Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ ÏôÑÎ£å: {df_final_train.shape}")
print(f"‚úÖ ÌõÑÎ≥¥ Î≥ÄÏàò Î¶¨Ïä§Ìä∏ ({len(final_selected)}Í∞ú): {final_selected}")

# ---------------------------------------------------------
# 3. Ï†ÑÌÉêÏÉâ (Exhaustive Search) - ÏµúÏ†ÅÏùò Î≥ÄÏàò Ï°∞Ìï© Ï∞æÍ∏∞
# ---------------------------------------------------------
print("\nüîç [Process 1] Ï†ÑÌÉêÏÉâ: ÏµúÏ†ÅÏùò Î≥ÄÏàò Ï°∞Ìï©(Subset) Î∞úÍµ¥ Ï§ë...")

def get_combinations(features, min_k=3, max_k=6):
    # ÌõÑÎ≥¥ Î≥ÄÏàò Ï§ë 3~6Í∞ú Ï°∞Ìï©ÏùÑ ÏÉùÏÑ±
    combi_list = []
    # ÏÜçÎèÑÎ•º ÏúÑÌï¥ ÏµúÎåÄ 8Í∞úÍπåÏßÄÎßå Ï°∞Ìï© ÎåÄÏÉÅÏóê Ìè¨Ìï® (Ïù¥ÎØ∏ Ï§ëÏöîÎèÑ Ïàú Ï†ïÎ†¨Îê®)
    target = features[:8]
    for k in range(min_k, max_k + 1):
        combi_list.extend(list(itertools.combinations(target, k)))
    return combi_list

def evaluate_subset_light(subset):
    # Í∞ÑÏù¥ Í≤ÄÏ¶ùÏö© Îç∞Ïù¥ÌÑ∞ÏÖã
    X_sub_raw = df_final_train[list(subset)].values
    y_sub_raw = df_final_train['Target_Next_Month'].values.reshape(-1, 1)

    # Ïä§ÏºÄÏùºÎü¨Îäî Ìä∏Î†àÏù∏ Îç∞Ïù¥ÌÑ∞ÏóêÎßå fit
    scaler_X_sub = StandardScaler()
    scaler_y_sub = StandardScaler()

    # ÏãúÌÄÄÏä§ ÏÉùÏÑ± (ÏõêÏ≤ú Îç∞Ïù¥ÌÑ∞ ÏÇ¨Ïö©)
    time_steps_sub = 6 # Í∞ÑÏù¥ Í≤ÄÏ¶ùÏùÑ ÏúÑÌïú Í≥†Ï†ï time_steps
    Xs_raw, ys_raw = [], []
    for i in range(len(X_sub_raw) - time_steps_sub):
        Xs_raw.append(X_sub_raw[i:(i+time_steps_sub)])
        ys_raw.append(y_sub_raw[i+time_steps_sub])
    Xs_raw, ys_raw = np.array(Xs_raw), np.array(ys_raw)

    # Train/Val Split (8:2) - Ïä§ÏºÄÏùºÎßÅ Ï†Ñ Ïä§ÌîåÎ¶ø
    split_sub = int(len(Xs_raw) * 0.8)
    X_train_sub_raw, X_val_sub_raw = Xs_raw[:split_sub], Xs_raw[split_sub:]
    y_train_sub_raw, y_val_sub_raw = ys_raw[:split_sub], ys_raw[split_sub:]

    # Ïä§ÏºÄÏùºÎü¨ fit_transformÏùÄ ÌïôÏäµ Îç∞Ïù¥ÌÑ∞ÏóêÎßå, transformÏùÄ Í≤ÄÏ¶ù Îç∞Ïù¥ÌÑ∞Ïóê
    X_train_sub_scaled = scaler_X_sub.fit_transform(X_train_sub_raw.reshape(-1, X_train_sub_raw.shape[-1])).reshape(X_train_sub_raw.shape)
    X_val_sub_scaled = scaler_X_sub.transform(X_val_sub_raw.reshape(-1, X_val_sub_raw.shape[-1])).reshape(X_val_sub_raw.shape)

    y_train_sub_scaled = scaler_y_sub.fit_transform(y_train_sub_raw)
    y_val_sub_scaled = scaler_y_sub.transform(y_val_sub_raw)

    model = Sequential([
        Input(shape=(time_steps_sub, len(subset))),
        LSTM(16, return_sequences=False),
        Dense(1)
    ])
    model.compile(optimizer=Adam(learning_rate=0.005), loss='mse')
    model.fit(X_train_sub_scaled, y_train_sub_scaled, epochs=10, batch_size=32, verbose=0, validation_data=(X_val_sub_scaled, y_val_sub_scaled))

    return model.evaluate(X_val_sub_scaled, y_val_sub_scaled, verbose=0)

# Ï°∞Ìï© ÌÉêÏÉâ Ïã§Ìñâ
subsets = get_combinations(final_selected)
if len(subsets) > 50:
    subsets = random.sample(subsets, 50) # ÎÑàÎ¨¥ ÎßéÏúºÎ©¥ 50Í∞ú ÏÉòÌîåÎßÅ

best_score = float('inf')
best_subset = []

for idx, subset in enumerate(subsets):
    score = evaluate_subset_light(subset)
    if score < best_score:
        best_score = score
        best_subset = subset
    if (idx+1) % 10 == 0:
        print(f"   -> ÏßÑÌñâÎ•† {idx+1}/{len(subsets)} | Current Best Loss: {best_score:.5f}")

print(f"‚úÖ ÏµúÏ†Å Ï°∞Ìï© ÌôïÏ†ï ({len(best_subset)}Í∞ú): {best_subset}")
final_features = list(best_subset)

# ---------------------------------------------------------
# 4. Îç∞Ïù¥ÌÑ∞ÏÖã Íµ¨Ï∂ï (Final Features Í∏∞Ï§Ä) - ÏõêÏ≤ú Îç∞Ïù¥ÌÑ∞ Ïú†ÏßÄ
# ---------------------------------------------------------
X_raw = df_final_train[final_features].values
y_raw = df_final_train['Target_Next_Month'].values.reshape(-1, 1)

def create_sequences(X, y, time_steps):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

# ---------------------------------------------------------
# 5. Quantile Loss & Attention Model
# ---------------------------------------------------------
def quantile_loss(q, y_true, y_pred):
    e = y_true - y_pred
    return tf.maximum(q * e, (q - 1) * e)

def combined_loss(y_true, y_pred):
    # Output: [Lower(0.025), Median(0.5), Upper(0.975)]
    l = quantile_loss(0.025, y_true, y_pred[:, 0:1])
    m = quantile_loss(0.5, y_true, y_pred[:, 1:2])
    u = quantile_loss(0.975, y_true, y_pred[:, 2:3])
    return tf.reduce_mean(l + m + u)

def create_attention_model(time_steps, n_features, units, dropout):
    inputs = Input(shape=(time_steps, n_features))
    # AttentionÏùÑ ÏúÑÌï¥ return_sequences=True
    lstm_out = LSTM(units, return_sequences=True)(inputs)
    lstm_out = Dropout(dropout)(lstm_out)

    # Self-Attention Mechanism
    # Query=Value=Key=lstm_out
    att = Attention()([lstm_out, lstm_out])

    # Global Average PoolingÏúºÎ°ú ÏãúÌÄÄÏä§ ÏïïÏ∂ï
    pool = GlobalAveragePooling1D()(att)

    x = Dense(16, activation='relu')(pool)
    outputs = Dense(3)(x) # 3 Quantiles

    return Model(inputs=inputs, outputs=outputs)

# ---------------------------------------------------------
# 6. Optuna ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ ÏµúÏ†ÅÌôî
# ---------------------------------------------------------
print("\nüîç [Process 2] Optuna ÏµúÏ†ÅÌôî ÏãúÏûë...")

def objective(trial):
    # Search Space
    ts = trial.suggest_int('time_steps', 3, 12)
    units = trial.suggest_categorical('units', [16, 32, 64])
    dr = trial.suggest_float('dropout', 0.1, 0.4)
    lr = trial.suggest_float('lr', 1e-4, 1e-2, log=True)
    bs = trial.suggest_categorical('batch_size', [16, 32])

    # Data Prep - Ïä§ÏºÄÏùºÎßÅ Ï†Ñ ÏãúÌÄÄÏä§ ÏÉùÏÑ± Î∞è Î∂ÑÌï†
    X_seq_raw, y_seq_raw = create_sequences(X_raw, y_raw, ts)
    split_point = int(len(X_seq_raw) * 0.8)
    X_tr_raw, X_val_raw = X_seq_raw[:split_point], X_seq_raw[split_point:]
    y_tr_raw, y_val_raw = y_seq_raw[:split_point], y_seq_raw[split_point:]

    # Time-series aware scaling for Optuna trials
    scaler_X_trial = StandardScaler()
    scaler_y_trial = StandardScaler()

    # Fit scalers only on training data and transform both training and validation
    X_tr_scaled = scaler_X_trial.fit_transform(X_tr_raw.reshape(-1, X_tr_raw.shape[-1])).reshape(X_tr_raw.shape)
    X_val_scaled = scaler_X_trial.transform(X_val_raw.reshape(-1, X_val_raw.shape[-1])).reshape(X_val_raw.shape)

    y_tr_scaled = scaler_y_trial.fit_transform(y_tr_raw)
    y_val_scaled = scaler_y_trial.transform(y_val_raw)

    # Model Build & Train
    model = create_attention_model(ts, len(final_features), units, dr)
    model.compile(optimizer=Adam(learning_rate=lr), loss=combined_loss)

    es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    history = model.fit(X_tr_scaled, y_tr_scaled, validation_data=(X_val_scaled, y_val_scaled),
                        epochs=50, batch_size=bs, callbacks=[es], verbose=0)

    return min(history.history['val_loss'])

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=15)

best_params = study.best_params
print(f"‚úÖ ÏµúÏ†Å ÌååÎùºÎØ∏ÌÑ∞ ÎèÑÏ∂ú: {best_params}")

# ---------------------------------------------------------
# 7. ÏµúÏ¢Ö Î™®Îç∏ ÌïôÏäµ (Full Data)
# ---------------------------------------------------------
print("\nüî• [Process 3] ÏµúÏ¢Ö Î™®Îç∏ ÌïôÏäµ (Attention LSTM)...")

final_ts = best_params['time_steps']

# ÏµúÏ¢Ö Î™®Îç∏ ÌïôÏäµÏùÑ ÏúÑÌïú Ïä§ÏºÄÏùºÎü¨ (Ï†ÑÏ≤¥ Í∞ÄÏö© Îç∞Ïù¥ÌÑ∞Ïóê fit)
# Ïù¥ Ïä§ÏºÄÏùºÎü¨Îì§ÏùÄ Ïù¥ÌõÑ ÏòàÏ∏° Î∞è ÏãúÍ∞ÅÌôîÏóêÎèÑ ÏÇ¨Ïö©ÎêòÎØÄÎ°ú Ï†ÑÏó≠ Î≥ÄÏàòÎ°ú ÏÑ†Ïñ∏
global scaler_X_final, scaler_y_final, X_scaled_final, y_scaled_final

scaler_X_final = StandardScaler()
scaler_y_final = StandardScaler()

X_scaled_final = scaler_X_final.fit_transform(X_raw) # Ï†ÑÏ≤¥ X_rawÏóê fit_transform
y_scaled_final = scaler_y_final.fit_transform(y_raw) # Ï†ÑÏ≤¥ y_rawÏóê fit_transform

X_final, y_final = create_sequences(X_scaled_final, y_scaled_final, final_ts)

final_model = create_attention_model(
    final_ts, len(final_features),
    best_params['units'], best_params['dropout']
)
final_model.compile(optimizer=Adam(learning_rate=best_params['lr']), loss=combined_loss)

final_model.fit(X_final, y_final, epochs=100, batch_size=best_params['batch_size'], verbose=1)

# ---------------------------------------------------------
# 8. 2025ÎÖÑ 9Ïõî ÏòàÏ∏° Î∞è Î≥µÏõê (Reconstruction)
# ---------------------------------------------------------
print("\nüîÆ ÏòàÏ∏° ÏàòÌñâ Ï§ë...")

# ÎßàÏßÄÎßâ ÏúàÎèÑÏö∞ Ï∂îÏ∂ú (2025-08 Ìè¨Ìï®)
# X_rawÏùò ÎßàÏßÄÎßâ Î∂ÄÎ∂ÑÏù¥ Í∞ÄÏû• ÏµúÍ∑º Îç∞Ïù¥ÌÑ∞ÏûÑ
last_seq_raw = X_raw[-final_ts:]
last_seq_sc = scaler_X_final.transform(last_seq_raw).reshape(1, final_ts, -1)

# ÏòàÏ∏° (Scaled dlog Îã®ÏúÑ)
pred_sc = final_model.predict(last_seq_sc)
pred_orig = scaler_y_final.inverse_transform(pred_sc) # [Lower, Median, Upper]

# ÌôòÏú® Î†àÎ≤® Î≥µÏõê: FX_t+1 = FX_t * exp(dlog)
last_fx = df_final_train['FX_Level'].iloc[-1]
last_date = df_final_train.index[-1]

forecast_vals = last_fx * np.exp(pred_orig[0])
lower, median, upper = forecast_vals

print("="*60)
print(f"üìÖ Îç∞Ïù¥ÌÑ∞ ÎßàÏßÄÎßâ Í∏∞Ï§ÄÏùº: {last_date.date()} (ÌôòÏú®: {last_fx:,.2f}Ïõê)")
print("-" * 60)
print(f"üéØ [2025ÎÖÑ 9Ïõî] USD/KRW ÌôòÏú® ÏòàÏ∏° Í≤∞Í≥º")
print(f"   üìâ Lower Bound (2.5%):  {lower:,.2f} Ïõê")
print(f"   üîµ Median Forecast    :  {median:,.2f} Ïõê")
print(f"   üìà Upper Bound (97.5%): {upper:,.2f} Ïõê")
print("="*60)
```

```python
# ---------------------------------------------------------
# [Step 4] Î™®Îç∏ ÏÑ±Îä• ÌèâÍ∞Ä (Metrics) Î∞è 9Ïõî ÏòàÏ∏° ÏãúÍ∞ÅÌôî (Fan Chart)
# ---------------------------------------------------------
# Ï£ºÏùò: Ïù¥ ÏΩîÎìúÎäî Step 3 Ïã§Ìñâ ÌõÑ, Î©îÎ™®Î¶¨Ïóê Î™®Îç∏(final_model)Í≥º
# Îç∞Ïù¥ÌÑ∞(df_final_train)Í∞Ä ÏûàÎäî ÏÉÅÌÉúÏóêÏÑú Ïã§ÌñâÌï¥Ïïº Ìï©ÎãàÎã§.

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (Ïù¥Ï†Ñ Îã®Í≥ÑÏóêÏÑú ÏÑ§Ï†ïÎêú font_propÏù¥ ÏûàÎã§Î©¥ ÏÇ¨Ïö©)
title_font = font_prop if 'font_prop' in locals() else None

# =========================================================
# üéöÔ∏è [ÏÇ¨Ïö©Ïûê ÏÑ§Ï†ï] Ïã†Î¢∞Íµ¨Í∞Ñ ÎÑàÎπÑ Ï°∞Ï†ï (Interval Scaling)
# =========================================================
# 1.0 = ÏõêÎûò Î™®Îç∏Ïùò 95% Íµ¨Í∞Ñ (Î≥¥ÏàòÏ†Å, ÎÑìÏùå)
# 0.7 ~ 0.8 = Ïã§Î¨¥Ï†Å 80% ÏàòÏ§Ä Íµ¨Í∞Ñ (Í≥µÍ≤©Ï†Å, Ï¢ÅÏùå) -> Ï∂îÏ≤ú
interval_scale = 0.75
print(f"\nüîß Ïã†Î¢∞Íµ¨Í∞Ñ Ï°∞Ï†ï Í≥ÑÏàò Ï†ÅÏö©: {interval_scale}Î∞∞ (Í∏∞Ï°¥ ÎåÄÎπÑ {100*(1-interval_scale):.0f}% Ï∂ïÏÜå)")
# =========================================================

print("\nüìä [1. Îç∞Ïù¥ÌÑ∞ ÏûÖÎ†• Ï∞®Ïõê ÏûêÎèô Î≥¥Ï†ï]")

# ---------------------------------------------------------
# 0. ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ Ï∞®Ïõê Î∂àÏùºÏπò Ìï¥Í≤∞ (Auto-Fix)
# ---------------------------------------------------------
try:
    expected_features = final_model.input_shape[-1]
    expected_timesteps = final_model.input_shape[1]
except AttributeError:
    print("‚ö†Ô∏è Î™®Îç∏(final_model)ÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§. Step 3Î•º Î®ºÏ†Ä Ïã§ÌñâÌï¥Ï£ºÏÑ∏Ïöî.")
    raise

# ÌòÑÏû¨ Î©îÎ™®Î¶¨Ïóê ÏûàÎäî ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞(X_raw)Ïùò Feature Í∞úÏàò ÌôïÏù∏
current_features = X_raw.shape[1] if 'X_raw' in locals() else 0

print(f"   üîπ Î™®Îç∏ Í∏∞ÎåÄ Feature: {expected_features}Í∞ú")
print(f"   üîπ ÌòÑÏû¨ ÏûÖÎ†• Feature: {current_features}Í∞ú")

# Ï∞®ÏõêÏù¥ ÎßûÏßÄ ÏïäÏúºÎ©¥ ÏûêÎèôÏúºÎ°ú Ïû¨Íµ¨ÏÑ±
if current_features != expected_features:
    print("   ‚ö†Ô∏è Ï∞®Ïõê Î∂àÏùºÏπò Í∞êÏßÄ! Î™®Îç∏Ïóê ÎßûÎäî ÏµúÏ†Å Î≥ÄÏàò Ï°∞Ìï©ÏúºÎ°ú Îç∞Ïù¥ÌÑ∞Î•º Ïû¨Íµ¨ÏÑ±Ìï©ÎãàÎã§.")

    if 'final_features' in locals() and len(final_features) == expected_features:
        print(f"   ‚úÖ ÏµúÏ†Å Î≥ÄÏàò Ï†ÅÏö©: {final_features}")

        # 1) Îç∞Ïù¥ÌÑ∞ Ïû¨Íµ¨Ï∂ï
        X_raw = df_final_train[final_features].values
        y_raw = df_final_train['Target_Next_Month'].values.reshape(-1, 1)

        # Ïù¥ ÏÖÄÏóêÏÑú ÌèâÍ∞ÄÎ•º ÏúÑÌï¥ Îã§Ïãú Ïä§ÏºÄÏùºÎü¨Î•º fit_transform Ìï®
        scaler_X_eval = StandardScaler()
        scaler_y_eval = StandardScaler()
        X_scaled_eval = scaler_X_eval.fit_transform(X_raw)
        y_scaled_eval = scaler_y_eval.fit_transform(y_raw)

        # 3) ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞ Îã§Ïãú ÏÉùÏÑ± Ìï®Ïàò
        def create_sequences_fix(X, y, time_steps):
            Xs, ys = [], []
            for i in range(len(X) - time_steps):
                Xs.append(X[i:(i + time_steps)])
                ys.append(y[i + time_steps])
            return np.array(Xs), np.array(ys)

        # 4) ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞(X_final_eval) Í∞±Ïã†
        X_final_eval, y_final_eval = create_sequences_fix(X_scaled_eval, y_scaled_eval, expected_timesteps)
        final_ts = expected_timesteps
        print("   ‚úÖ Îç∞Ïù¥ÌÑ∞ Î∞è Ïä§ÏºÄÏùºÎü¨ Ïû¨Íµ¨ÏÑ± ÏôÑÎ£å.")
    else:
        raise ValueError(f"Î™®Îç∏ÏùÄ {expected_features}Í∞ú Î≥ÄÏàòÎ•º ÏõêÌïòÎäîÎç∞, Îß§Ïπ≠ÎêòÎäî 'final_features'Î•º Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§. Step 3Î•º Îã§Ïãú Ïã§ÌñâÌï¥Ï£ºÏÑ∏Ïöî.")
else:
    print("   ‚úÖ ÏûÖÎ†• Ï∞®ÏõêÏù¥ Ï†ïÏÉÅÏûÖÎãàÎã§.")
    if 'final_ts' not in locals():
        final_ts = expected_timesteps

    # Step 3ÏóêÏÑú ÌïôÏäµÎêú Ïä§ÏºÄÏùºÎü¨ÏôÄ Ïä§ÏºÄÏùºÎßÅÎêú Îç∞Ïù¥ÌÑ∞Î•º Ïû¨ÏÇ¨Ïö©
    if 'scaler_X_final' not in locals() or 'scaler_y_final' not in locals() or \
       'X_scaled_final' not in locals() or 'y_scaled_final' not in locals():
        raise ValueError("‚ö†Ô∏è Step 3ÏóêÏÑú ÏµúÏ¢Ö Ïä§ÏºÄÏùºÎü¨(scaler_X_final, scaler_y_final)ÏôÄ Ïä§ÏºÄÏùºÎßÅÎêú Îç∞Ïù¥ÌÑ∞Í∞Ä Ï†ïÏùòÎêòÏßÄ ÏïäÏïòÏäµÎãàÎã§. Step 3Î•º Îã§Ïãú Ïã§ÌñâÌï¥Ï£ºÏÑ∏Ïöî.")

    scaler_X_eval = scaler_X_final
    scaler_y_eval = scaler_y_final
    X_scaled_eval = X_scaled_final
    y_scaled_eval = y_scaled_final

    def create_sequences_fix(X, y, time_steps):
        Xs, ys = [], []
        for i in range(len(X) - time_steps):
            Xs.append(X[i:(i + time_steps)])
            ys.append(y[i + time_steps])
        return np.array(Xs), np.array(ys)

    X_final_eval, y_final_eval = create_sequences_fix(X_scaled_eval, y_scaled_eval, final_ts)


# ---------------------------------------------------------
# 1. Í≤ÄÏ¶ùÏö© Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ (Train/Test Split)
# ---------------------------------------------------------
print("\nüìä [2. Î™®Îç∏ ÏÑ±Îä• ÌèâÍ∞Ä ÏßÄÌëú]")

# Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞Ïùò Îí§Ï™Ω 10%Î•º ÌÖåÏä§Ìä∏ÏÖãÏúºÎ°ú ÏÇ¨Ïö©
split_idx = int(len(X_final_eval) * 0.9)

# [Í≥ºÏ†ÅÌï© ÌôïÏù∏Ïö©] Train Set
X_train_eval = X_final_eval[:split_idx]
y_train_eval = y_final_eval[:split_idx]

# Test Set
X_test_eval = X_final_eval[split_idx:]
y_test_eval = y_final_eval[split_idx:]

# ÏãúÍ∞ÅÌôîÎ•º ÏúÑÌïú ÎÇ†Ïßú Ïù∏Îç±Ïä§ Ï∂îÏ∂ú
test_len = len(y_test_eval)
test_indices = df_final_train.index[-test_len:]
train_indices = df_final_train.index[final_ts:split_idx+final_ts]


# ---------------------------------------------------------
# 2. ÏÑ±Îä• ÏßÄÌëú Í≥ÑÏÇ∞ Î∞è Í≥ºÏ†ÅÌï© ÌôïÏù∏
# ---------------------------------------------------------
# (1) Test Set ÏòàÏ∏°
pred_test_sc = final_model.predict(X_test_eval, verbose=0)
pred_test_dlog = scaler_y_eval.inverse_transform(pred_test_sc)
y_test_dlog = scaler_y_eval.inverse_transform(y_test_eval)

# KRW Î†àÎ≤® Î≥µÏõê
prev_fx_test = df_final_train.loc[test_indices, 'FX_Level'].values
fx_pred_test = prev_fx_test * np.exp(pred_test_dlog[:, 1]) # Median
fx_true_test = prev_fx_test * np.exp(y_test_dlog.flatten())

# Test Metrics
rmse_test_krw = np.sqrt(mean_squared_error(fx_true_test, fx_pred_test))
mae_test_krw = mean_absolute_error(fx_true_test, fx_pred_test)

# (2) Train Set ÏòàÏ∏°
print("   ... ÌïôÏäµ Îç∞Ïù¥ÌÑ∞(Train Set) ÏÑ±Îä• Í≥ÑÏÇ∞ Ï§ë ...")
pred_train_sc = final_model.predict(X_train_eval, verbose=0)
pred_train_dlog = scaler_y_eval.inverse_transform(pred_train_sc)
y_train_dlog = scaler_y_eval.inverse_transform(y_train_eval)

prev_fx_train = df_final_train.loc[train_indices, 'FX_Level'].values
fx_pred_train = prev_fx_train * np.exp(pred_train_dlog[:, 1])
fx_true_train = prev_fx_train * np.exp(y_train_dlog.flatten())
rmse_train_krw = np.sqrt(mean_squared_error(fx_true_train, fx_pred_train))

# (3) Í≤∞Í≥º Ï∂úÎ†•
print("-" * 50)
print("üìå [ÏÑ±Îä• ÎπÑÍµê] Train(ÌïôÏäµ) vs Test(Í≤ÄÏ¶ù)")
print(f"   üîπ Train RMSE : {rmse_train_krw:.2f} Ïõê")
print(f"   üîπ Test  RMSE : {rmse_test_krw:.2f} Ïõê")
print(f"   üëâ Ï∞®Ïù¥(Gap)  : {rmse_test_krw - rmse_train_krw:.2f} Ïõê")
print("-" * 50)


# ---------------------------------------------------------
# [Ï∂îÍ∞Ä] Í≥ºÏ†ÅÌï© Í≤ÄÏ¶ù ÏãúÍ∞ÅÌôî (Train vs Test Plot)
# ---------------------------------------------------------
print("\nüìà [1. Í≥ºÏ†ÅÌï© Í≤ÄÏ¶ù ÏãúÍ∞ÅÌôî]")
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

axes[0].plot(train_indices, fx_true_train, 'k-', label='Actual', linewidth=1)
axes[0].plot(train_indices, fx_pred_train, 'b--', label='Predicted (Train)', linewidth=1)
axes[0].set_title(f'Train Set Performance (RMSE: {rmse_train_krw:.1f})', fontproperties=title_font)
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(test_indices, fx_true_test, 'k-', label='Actual', linewidth=1.5)
axes[1].plot(test_indices, fx_pred_test, 'r--', label='Predicted (Test)', linewidth=1.5)
axes[1].set_title(f'Test Set Performance (RMSE: {rmse_test_krw:.1f})', fontproperties=title_font)
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()


# ---------------------------------------------------------
# 3. 2025ÎÖÑ 9Ïõî ÏµúÏ¢Ö ÏòàÏ∏° Î∞è Ïã†Î¢∞Íµ¨Í∞Ñ Ï°∞Ï†ï (Interval Scaling)
# ---------------------------------------------------------
# ÎßàÏßÄÎßâ ÏûÖÎ†• ÏãúÌÄÄÏä§
last_seq_raw = X_raw[-final_ts:]
last_seq_sc = scaler_X_eval.transform(last_seq_raw).reshape(1, final_ts, -1)

# ÏòàÏ∏° ÏàòÌñâ
future_pred_sc = final_model.predict(last_seq_sc, verbose=0)
future_pred_dlog = scaler_y_eval.inverse_transform(future_pred_sc) # [Lower, Median, Upper]

# Î†àÎ≤® Î≥µÏõê (Raw Scale)
last_fx = df_final_train['FX_Level'].iloc[-1]
last_date = df_final_train.index[-1]
target_date = last_date + pd.DateOffset(months=1) # 2025-09

# [ÌïµÏã¨] Ïã†Î¢∞Íµ¨Í∞Ñ ÎÑàÎπÑ Ï°∞Ï†ï Î°úÏßÅ (Scaling)
# MedianÏùÑ Ï§ëÏã¨ÏúºÎ°ú ÏúÑ/ÏïÑÎûò Í∞ÑÍ≤©ÏùÑ ÏÇ¨Ïö©ÏûêÍ∞Ä ÏÑ§Ï†ïÌïú ÎπÑÏú®(interval_scale)ÎßåÌÅº Ï§ÑÏûÖÎãàÎã§.
raw_lower_dlog = future_pred_dlog[0][0]
raw_median_dlog = future_pred_dlog[0][1]
raw_upper_dlog = future_pred_dlog[0][2]

# Ï°∞Ï†ïÎêú dlog Í≥ÑÏÇ∞
adj_lower_dlog = raw_median_dlog - (raw_median_dlog - raw_lower_dlog) * interval_scale
adj_upper_dlog = raw_median_dlog + (raw_upper_dlog - raw_median_dlog) * interval_scale

# ÏµúÏ¢Ö ÌôòÏú® ÌôòÏÇ∞
pred_lower = last_fx * np.exp(adj_lower_dlog)
pred_median = last_fx * np.exp(raw_median_dlog) # MedianÏùÄ Í∑∏ÎåÄÎ°ú
pred_upper = last_fx * np.exp(adj_upper_dlog)

print(f"\nüîÆ [ÏµúÏ¢Ö Í≤∞Í≥º] 2025ÎÖÑ 9Ïõî ÏòàÏ∏° ({target_date.strftime("%Y-%m")})")
print(f"   (Ï†ÅÏö©Îêú Íµ¨Í∞Ñ Ïä§ÏºÄÏùº: {interval_scale}Î∞∞)")
print(f"   üìâ Lower  : {pred_lower:,.2f} Ïõê")
print(f"   üîµ Median : {pred_median:,.2f} Ïõê")
print(f"   üìà Upper  : {pred_upper:,.2f} Ïõê")


# ---------------------------------------------------------
# 4. 9Ïõî ÏòàÏ∏° Ï†ÑÏö© ÏãúÍ∞ÅÌôî (Fan Chart Update)
# ---------------------------------------------------------
print("\nüìà [2. ÏµúÏ¢Ö ÏòàÏ∏° Fan Chart]")

# (A) Ï†ÑÏ≤¥ Íµ¨Í∞Ñ Î™®Îç∏ ÏòàÏ∏°Í∞í (Backtest Line)
all_pred_sc = final_model.predict(X_final_eval, verbose=0)
all_pred_dlog = scaler_y_eval.inverse_transform(all_pred_sc)

# Ï†ÑÏ≤¥ Í∏∞Í∞Ñ Ïù∏Îç±Ïä§
full_indices = df_final_train.index[final_ts:]
prev_fx_full = df_final_train.loc[full_indices, 'FX_Level'].values
full_pred_krw = prev_fx_full * np.exp(all_pred_dlog[:, 1]) # Median

# (B) Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ Î≤îÏúÑ ÏÑ§Ï†ï (2024ÎÖÑ 1ÏõîÎ∂ÄÌÑ∞)
start_plot_date = '2024-01-01'
try:
    plot_start_idx = df_final_train.index.get_loc(pd.Timestamp(start_plot_date))
except:
    plot_start_idx = -24

history_dates = df_final_train.index[plot_start_idx:]
history_vals = df_final_train['FX_Level'].iloc[plot_start_idx:]

# Î™®Îç∏ ÏòàÏ∏°Í∞í Ïä¨ÎùºÏù¥Ïã±
model_pred_series = pd.Series(full_pred_krw, index=full_indices)
try:
    history_model_vals = model_pred_series.loc[start_plot_date:]
except:
    history_model_vals = model_pred_series.iloc[-24:]

plt.figure(figsize=(12, 7))

# (1) Í≥ºÍ±∞ Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞
plt.plot(history_dates, history_vals, 'k-', label='Historical Actual', linewidth=2)

# (2) Í≥ºÍ±∞ Î™®Îç∏ ÏòàÏ∏° Îç∞Ïù¥ÌÑ∞ (Model Fit)
plt.plot(history_model_vals.index, history_model_vals.values, 'b--', label='Model Fit (Past)', linewidth=1.5, alpha=0.7)

# (3) ÎØ∏Îûò ÏòàÏ∏° Ïó∞Í≤∞
plt.plot([last_date, target_date], [last_fx, pred_median], 'r--', label='Future Forecast', linewidth=2)

# (4) Fan Chart (Ï°∞Ï†ïÎêú Ïã†Î¢∞Íµ¨Í∞Ñ)
plt.fill_between(
    [last_date, target_date],
    [last_fx, pred_lower],
    [last_fx, pred_upper],
    color='red', alpha=0.2, label=f'Confidence Interval (Scaled {interval_scale}x)'
)

# (5) ÌÖçÏä§Ìä∏ Ï£ºÏÑù
text_offset = (pred_upper - pred_lower) * 0.1 # ÌÖçÏä§Ìä∏ Í∞ÑÍ≤© ÌôïÎ≥¥
plt.text(target_date, pred_upper + text_offset, f'{pred_upper:,.0f}', color='darkred', va='bottom', ha='left', fontweight='bold')
plt.text(target_date, pred_median, f'{pred_median:,.0f}', color='red', va='center', ha='left', fontweight='bold')
plt.text(target_date, pred_lower - text_offset, f'{pred_lower:,.0f}', color='darkred', va='top', ha='left', fontweight='bold')

plt.title(f'USD/KRW Forecast ({target_date.strftime("%Y-%m")})', fontsize=16, fontproperties=title_font)
plt.ylabel('Exchange Rate (KRW)')
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(loc='upper left')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
